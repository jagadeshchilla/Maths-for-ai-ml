### 1. Univariate Analysis (1 Variable)
Purpose: Examine one variable to understand distribution, spread, and frequency.

**A) Numerical Variable**

Suggested Plots: Histogram, Boxplot, Line Chart

Dataset: tips dataset from Seaborn (total_bill column)

1. Histogram

Definition: Shows the frequency distribution of numeric data.

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips=sns.load_dataset('tips')
data=tips['total_bill']
```

```python
data.head()
```

```python
# Histogram
plt.figure(figsize=(8,4))
sns.histplot(data, bins=15, kde=True, color='skyblue')
plt.title("Histogram of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Frequency")
plt.show()
```

Explanation:

- X-axis → total bill amount

- Y-axis → number of bills in each range

- KDE curve → smooth estimate of distribution

2. Boxplot

Definition: Visualizes spread, median, quartiles, and outliers.

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=data, color='lightgreen')
plt.title("Boxplot of Total Bill")
plt.xlabel("Total Bill ($)")
plt.show()
```

Explanation:

- Box → Q1 to Q3 (middle 50%)

- Line inside box → median

- Whiskers → range excluding outliers

- Points outside whiskers → outliers

3. Line Chart (for trend / ordered data)

Definition: Shows trend or pattern over an ordered sequence.

Example: Daily sales over a week (generated data)

```python
import numpy as np

# Generated daily sales data
days = ['Mon','Tue','Wed','Thu','Fri','Sat','Sun']
sales = np.random.randint(50, 150, size=7)

plt.figure(figsize=(8,4))
plt.plot(days, sales, marker='o', linestyle='-', color='purple')
plt.title("Daily Ice Cream Sales")
plt.xlabel("Day")
plt.ylabel("Number Sold")
plt.show()
```

Explanation:

- X-axis → days of the week

- Y-axis → number of ice creams sold

- Line shows daily sales trend

B) Categorical Variable

Suggested Plots: Bar Chart, Pie Chart

Dataset: tips dataset (day column)

1. Bar Chart

Definition: Shows frequency of each category.

```python
plt.figure(figsize=(6,4))
sns.countplot(x='day', data=tips, palette='pastel')
plt.title("Number of Bills by Day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

2. Pie Chart

Definition: Shows proportion of each category.

```python
day_counts = tips['day'].value_counts()
plt.figure(figsize=(6,6))
plt.pie(day_counts, labels=day_counts.index, autopct='%1.1f%%', colors=['skyblue','lightgreen','orange','pink'])
plt.title("Proportion of Bills by Day")
plt.show()
```

Explanation:

- Each slice → proportion of bills

- Shows percentage share visually

### 2. Bivariate Analysis (2 Variables)
Purpose: Examine the relationship between two variables.

**A) Numeric vs Numeric**

Suggested Plots: Scatter Plot, Line Plot, Hexbin Plot

Dataset: mpg dataset from Seaborn (miles per gallon vs horsepower)

1. Scatter Plot

Definition: Shows relationship/correlation between two numeric variables.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load dataset
mpg = sns.load_dataset("mpg").dropna()
x = mpg['horsepower']
y = mpg['mpg']
```

```python
# Scatter plot
plt.figure(figsize=(8,5))
sns.scatterplot(x=x, y=y, color='blue')
plt.title("MPG vs Horsepower")
plt.xlabel("Horsepower")
plt.ylabel("Miles per Gallon (MPG)")
plt.show()
```

Explanation:

- Negative trend → as horsepower increases, MPG decreases

- Each point → one car

2. Line Plot (if data is ordered)

Example: Generated time series for temperature vs ice cream sales

```python
import numpy as np

days=np.arange(1,11)
temperature=np.random.randint(25,35,size=10)
sales=temperature*10+np.random.randint(-5,5,size=10)
```

```python
plt.figure(figsize=(8,4))
plt.plot(temperature, sales, marker='o', linestyle='-', color='green')
plt.title("Ice Cream Sales vs Temperature")
plt.xlabel("Temperature (°C)")
plt.ylabel("Number Sold")
plt.show()
```

Explanation:

- Shows positive correlation → higher temperature → higher sales

3. Hexbin Plot (for dense data)

```python
plt.figure(figsize=(8,5))
plt.hexbin(x, y, gridsize=20, cmap='Blues')
plt.colorbar(label='Count')
plt.xlabel("Horsepower")
plt.ylabel("MPG")
plt.title("Hexbin: MPG vs Horsepower")
plt.show()
```

Explanation:

- Darker cells → more cars in that range

- Useful for large datasets

**B) Categorical vs Categorical**

Suggested Plots: Stacked Bar Chart, Mosaic Plot

Dataset: tips dataset (day vs smoker)

```python
import pandas as pd

# Stacked Bar Chart
ct = pd.crosstab(tips['day'], tips['smoker'])
ct.plot(kind='bar', stacked=True, color=['skyblue','orange'], figsize=(7,5))
plt.title("Smoker vs Day")
plt.xlabel("Day")
plt.ylabel("Count")
plt.show()
```

Explanation:

- Shows distribution of smokers and non-smokers across days

- Stacked bars → visual comparison of categories

**C) Categorical vs Numeric**

Suggested Plots: Boxplot, Violin Plot, Bar Plot with Mean

Dataset: tips dataset (day vs total_bill)

1. Boxplot

```python
plt.figure(figsize=(7,5))
sns.boxplot(x='day', y='total_bill', data=tips, palette='pastel')
plt.title("Total Bill Distribution by Day")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.show()
```

Explanation:

- Compare spread of total bills across days

- Outliers are visible

2. Violin Plot (distribution + density)

```python
plt.figure(figsize=(7,5))
sns.violinplot(x='day', y='total_bill', data=tips, palette='muted')
plt.title("Total Bill Distribution by Day (Violin)")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.show()
```

Explanation:

- Shows distribution shape + median

- Wider sections → more observations

3. Bar Plot with Mean

```python
plt.figure(figsize=(7,5))
sns.barplot(x='day', y='total_bill', data=tips, palette='coolwarm')
plt.title("Average Total Bill by Day")
plt.xlabel("Day")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

Explanation:

- Height of bars → average total bill per day

- Quick comparison of central tendency across categories

### Multivariate Analysis (3+ Variables)

Purpose: Examine complex relationships between multiple variables.

**A) Numerical + Numerical + Categorical**

Suggested Plots: 3D Scatter Plot, Faceted Scatter Plot, Bubble Chart

Dataset: penguins dataset from Seaborn

```python
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

penguins = sns.load_dataset("penguins").dropna()
```

```python
penguins.head()
```

```python
# 3D Scatter Plot: bill_length vs bill_depth vs species
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

species_colors = {'Adelie':'red', 'Chinstrap':'green', 'Gentoo':'blue'}
colors = penguins['species'].map(species_colors)

ax.scatter(penguins['bill_length_mm'], penguins['bill_depth_mm'], penguins['flipper_length_mm'], c=colors)
ax.set_xlabel('Bill Length (mm)')
ax.set_ylabel('Bill Depth (mm)')
ax.set_zlabel('Flipper Length (mm)')
ax.set_title("3D Scatter: Penguins Measurements by Species")
plt.show()
```

Explanation:

- X, Y, Z → numeric variables

- Color → categorical variable (species)

- Shows relationships between 3 numeric features and species

Bubble Chart (Numeric + Numeric + Categorical Size)

```python
plt.figure(figsize=(8,5))
plt.scatter(
    penguins['bill_length_mm'], penguins['flipper_length_mm'],
    s=penguins['body_mass_g']/50,  # bubble size proportional to body mass
    c=penguins['species'].map(species_colors), alpha=0.6
)
plt.xlabel('Bill Length (mm)')
plt.ylabel('Flipper Length (mm)')
plt.title("Bubble Chart: Bill vs Flipper with Body Mass")
plt.show()
```

Explanation:

- Bubble size → body mass (numeric)

- Color → species (categorical)

- Combines 3 numeric + 1 categorical variable

**B) Categorical + Categorical + Numerical**

Suggested Plots: Grouped Boxplot, Heatmap

Dataset: tips dataset (day + smoker + total_bill)

```python
# Grouped Boxplot
plt.figure(figsize=(8,5))
sns.boxplot(x='day', y='total_bill', hue='smoker', data=tips, palette='coolwarm')
plt.title("Total Bill by Day and Smoker Status")
plt.xlabel("Day")
plt.ylabel("Total Bill ($)")
plt.show()
```

Explanation:

- X-axis → day

- Hue → smoker status

- Boxplot → distribution of total bill per group

Heatmap (Categorical + Categorical + Numerical)

```python
ct = tips.pivot_table(values='total_bill', index='day', columns='smoker', aggfunc='mean')
sns.heatmap(ct, annot=True, cmap='YlGnBu', fmt=".1f")
plt.title("Average Total Bill: Day vs Smoker")
plt.show()
```

Explanation:

- Cells → mean total bill

- X-axis → smoker, Y-axis → day

- Heatmaps → quickly compare aggregated numeric values across categorical variables

**C) All Numerical Variables**

Suggested Plots: Correlation Heatmap, Pairplot

Dataset: penguins numeric features

```python
numeric_features = penguins[['bill_length_mm','bill_depth_mm','flipper_length_mm','body_mass_g']]
sns.heatmap(numeric_features.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap: Penguin Numeric Features")
plt.show()
```

Explanation:

- Shows correlation between all numeric variables

- Quick insight for feature selection, multicollinearity, or trends

Pairplot

```python
sns.pairplot(penguins, hue='species', vars=['bill_length_mm','bill_depth_mm','flipper_length_mm'])
plt.suptitle("Pairplot of Penguin Numeric Features by Species", y=1.02)
plt.show()
```

Explanation:

- Pairwise scatterplots + histogram diagonals

- Color → species

- Great for exploring multiple numeric relationships at once
